```python
import matplotlib.pyplot as plt

# Define the function and its derivative
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# Gradient descent function to find the minimum
def gradient_descent(function, derivative, x_start, learning_rate, epochs):
    x = x_start
    for _ in range(epochs):
        gradient = derivative(x)
        x = x - learning_rate * gradient  # Update x using the gradient
    y = function(x)  # Calculate y at the obtained x
    return x, y  # Return both x and y at the local minima

# Initial parameters
initial_x = 2
learning_rate = 0.1
num_epochs = 100

# Find the local minima using gradient descent
local_minima_x, local_minima_y = gradient_descent(function, derivative, initial_x, learning_rate, num_epochs)

print(f"Local Minima at x = {local_minima_x}, y = {local_minima_y}")

# Create the plot
def plot_gradient_descent(function, derivative, x_start, learning_rate, epochs):
    x = x_start
    x_values = [x]  # Store x values for plotting
    for _ in range(epochs):
        gradient = derivative(x)
        x = x - learning_rate * gradient  # Update x using the gradient
        x_values.append(x)  # Append the updated x for plotting
    
    return x_values

x_values = plot_gradient_descent(function, derivative, initial_x, learning_rate, num_epochs)

plt.figure(figsize=(8, 6))
plt.plot(range(len(x_values)), x_values, marker='o', linestyle='-', color='b')
plt.xlabel('Iterations')
plt.ylabel('Value of x')
plt.title('Gradient Descent: Value of x over Iterations')
plt.grid(True)
plt.show()
```
